## DIY 1

```python
# Trying to see how powerful the enumerative method is 
## by testing how large a number i could be before a pause in 
## printing can be noticed

max_val = int(input('Enter a positive integer:'))
i = 0
while i < max_val:
    i = i + 1
print(i)

import time

start_time = time.perf_counter() # Start the timer
end_time = time.perf_counter() # End the timer
elapsed_time = end_time - start_time # Calculate the elapsed time

print(f"Elapsed time: {elapsed_time:.6f} seconds")
```

## DIY 2

```python
## Test if an int > 2 is prime. If not, print largest divisor
x = int(input('Enter an integer greater than 2:'))
smallest_divisor = None
for guess in range(2,x):
    if x%guess == 0:
        smallest_divisor = guess
        largest_divisor = x/smallest_divisor
        break
if smallest_divisor != None:
    print('Largest divisor of', x, 'is', largest_divisor)
else:
    print(x, 'is a prime number')
```

## DIY 3

```python
def prime(num): # Function to check if a number is prime
    if num < 2:
        return True
    for i in range(3, int(num**0.5) + 1, 2):
        if num % i == 0:
            return False
    return True
sum_of_prime = 0
for num in range(3,1000,2):
    if prime(num):
        sum_of_prime += num
print("The sum of prime number that is greater than 2 and less than 1000 is:", sum_of_prime)
```

## DIY 4

```python
x = 123456 # Let's try to use more educated guesses
## knowing by constantly trying to find the middle we can
## get closer to the square faster,
## checkout the number of tries 30 instead of 3513631 (a 120,000 times improvement) !!)
epsilon = 0.01
num_guesses, low = 0,0
high = max(1,x)
ans = (high + low)/2
while abs(ans**2 - x) >= epsilon:
    print('low =', low, 'high = ', high, 'ans =', ans)
    num_guesses += 1
    if ans**2 < x:
        low = ans
    else:
        high = ans
    ans = (high + low)/2
print('number of guesses =', num_guesses)
print(ans, 'is close to square root of', x)

import matplotlib.pyplot as plt
# For x = 1,12,123,1234,12345,123456

x_values = [1,2,3,4,5,6]
y_values = [7,10,15,21,26,30]
print(x_values)
print(y_values)

plt.plot(x_values, y_values)

plt.xlabel("Digits")
plt.ylabel("Guesses")
plt.title("Number of digits vs. Number of guesses for different numbers")
plt.show()
```
